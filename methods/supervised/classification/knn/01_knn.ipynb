```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
X,y = make_moons(n_samples=100, noise=0.3,random_state=0)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
print(X.shape); print(y.shape)
```

```python
from collections import Counter
Counter(y)
```

```python
print(X_train.shape); print(X_test.shape); print(y_train.shape); print(y_test.shape)
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=y_test,s=50,cmap='viridis')
plt.title('Original class')
plt.show()
```

```python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Acc:", round(accuracy_score(y_test,y_pred),2))
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=y_pred,s=50,cmap='viridis')
plt.title('Predicted class')
plt.show()
```

```python
# find misclassified samples
misclassified_indices = np.where(y_test != y_pred)
misclassified_indices
```

```python
misclassified_indices[0]
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=y_pred,s=50,cmap='viridis')
# mark misclassified samples
plt.scatter(X_test[misclassified_indices,0],X_test[misclassified_indices,1],
            c='black',marker='X',s=200)
plt.title('Misclassified samples')
plt.show()
```
